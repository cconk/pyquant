```python
from thetadata import ThetaClient, StreamMsg, StreamMsgType

def streaming():
    # We are reading from a file that contains our thetadata credentials. You should manually add yours here.
    with open("credentials.txt") as file:
        client = ThetaClient(username=file.readline().rstrip(), passwd=file.readline().rstrip())

    client.connect_stream(callback)  # You can stop streaming by calling client.close_stream
    client.req_full_trade_stream_opt()  # requests every option trade (available to PRO tier subscribers)

# User generated method that gets called each time a message from the stream arrives.
def callback(msg: StreamMsg): 
    if msg.type == StreamMsgType.TRADE:
        print('---------------------------------------------------------------------------')
        print('contract:   ' + msg.contract.to_string()) # The contract
        print('trade:      ' + msg.trade.to_string())    # The trade
        print('ohlcvc:     ' + msg.ohlcvc.to_string())   # Open, high, low, close, volume, count
        print('last quote: ' + msg.quote.to_string())    # Last quote of the option at the time of trade.


if __name__ == "__main__":
    streaming()
```
