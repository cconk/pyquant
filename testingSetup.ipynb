```python
from openbb_terminal.sdk import openbb
import quantstats as qs
```

```python
data = openbb.stocks.load("AAPL", start_date="2012-06-01", end_date="2022-06-30")
```

```python
data.Close.plot()
```

```python
data.Close.rolling(30).mean().plot()
```

```python
aapl_returns = data['Adj Close'].pct_change()
qs.reports.metrics(
    aapl_returns,
    mode="full"
)
```

```python
from ibapi.client import EClient
from ibapi.wrapper import EWrapper

class IBapi(EWrapper, EClient):
    def __init__(self):
        EClient.__init__(self, self)

host = '127.0.0.1'
port = 7497

print(f'\nConnect {host}:{str(port)}')

app = IBapi()
app.connect(host, port, clientId=1)
app.run()
```
